## IMPORT LIBRARY

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## LOAD DATA

```python
df = pd.read_excel("US Superstore data.xls", engine="xlrd")
df.head()
```

# VISUALISASI

```python
# Menghitung total penjualan berdasarkan segment pelanggan
seg_profit = df.groupby('Segment')['Sales'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Segment', y='Sales', data=seg_profit)
plt.title('Total Sales by Segment')
plt.show()

# Menghitung rata-rata penjualan berdasarkan segment pelanggan
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Segment', y='Sales')
plt.title('Sales Distribution by Segment')
plt.xticks(rotation=45)
plt.show()
```

- Segmen pelanggan dengan total penjualan tertinggi adalah Consumer yang diikuti oleh Corporate dan terendah adalah Home Office.
- Meski total penjualan segmen Consumer tertinggi, rata-rata penjualannya hampir sama dengan segmen Corporate dan Home Office. Hal ini menunjukkan distribusi penjualan yang lebih merata antar kategori.

```python
# Menghitung total keuntungan berdasarkan negara bagian pelanggan
seg_profit = df.groupby('Region')['Profit'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Profit', data=seg_profit)
plt.title('Total Profit by Region')
plt.show()

# Menghitung rata-rata keuntungan berdasarkan negara bagian pelanggan
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Region', y='Profit')
plt.title('Profit Distribution by Region')
plt.xticks(rotation=45)
plt.show()
```

- Wilayah dengan total keuntungan tertinggi adalah West yang diikuti oleh East, South, dan Central.
- Berdasarkan negara bagian, urutan dari tertinggi ke terendah rata-rata keuntungan dan total keuntungannya sama.

```python
# Menghitung total profit berdasarkan wilayah pelanggan
seg_profit = df.groupby('Category')['Profit'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Profit', data=seg_profit)
plt.title('Total Profit by Category')
plt.show()

# Menghitung rata-rata keuntungan berdasarkan wilayah pelanggan
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Category', y='Profit')
plt.title('Profit Distribution by Category')
plt.xticks(rotation=45)
plt.show()
```

- Kategori dengan total keuntungan tertinggi adalah Technology yang diikuti oleh Office Supplies dan terendah adalah Furniture.
- Pada Office Supplies, keuntungan besar berasal dari volume transaksi yang sangat tinggi, meskipun setiap transaksi secara individual memberikan margin profit yang kecil.

```python
# Menghitung total penjualan berdasarkan mode pengiriman
sales_by_ship_mode = df.groupby('Ship Mode')['Sales'].sum()

plt.figure(figsize=(10, 6))
sales_by_ship_mode.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'gold', 'lightgreen', 'blue'], legend=False)
plt.title('Sales Distribution by Ship Mode')
plt.ylabel('')
plt.show()
```

- Pada Ship Mode, sebagian besar penjualan menggunakan pengiriman Standard Class (59,1%) dengan sisanya menggunakan Second Class (15,3%), First Class (20%), dan Same Day (5,6%).

```python
# Menghitung total keuntungan berdasarkan kategori
profit_by_category = df.groupby('Category')['Profit'].sum()

plt.figure(figsize=(10, 6))
profit_by_category.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'gold', 'lightgreen', 'blue'], legend=False)
plt.title('Sales Distribution by Category')
plt.ylabel('')
plt.show()
```

- Pada Kategori, penjualan terbanyak berasal dari Technology (50,8%) dengan sisanya menggunakan Office Supplies (42,8%) dan Furniture (6,4%).

```python
# Menyaring dan menjumlahkan penjualan per bulan
monthly_profit = df.groupby(df['Order Date'].dt.to_period('M'))['Profit'].sum()

plt.figure(figsize=(10, 6))
monthly_profit.plot(kind='line', marker='o', color='b')
plt.title('Monthly Profit')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

- Berdasarkan grafik, keuntungan tertinggi berada pada bulan Desember 2016 dan keuntungan terendah berada pada Januari 2015.
