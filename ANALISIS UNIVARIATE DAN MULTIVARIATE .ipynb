- Row ID menunjukkan distribusi yang merata, mengindikasikan bahwa data ini adalah pengidentifikasi unik tanpa pola tertentu.

- Sales menunjukkan distribusi yang sangat condong ke kiri (right-skewed) dengan sebagian besar nilai berada di kisaran rendah tetapi ada beberapa nilai yang sangat tinggi.
- Quantity memiliki distribusi yang tidak merata dengan sebagian besar transaksi mencatat kuantitas kecil dan sedikit yang mencatat kuantitas besar.
- Discount menunjukkan dua puncak yang menonjol di dekat nol dan nilai diskon yang umum digunakan.
- Profit menunjukkan distribusi yang simetris dengan puncak di sekitar nol, mengindikasikan bahwa sebagian besar transaksi memiliki keuntungan kecil atau nol, dengan beberapa mengalami kerugian besar atau keuntungan besar.

## ANALISIS MULTIVARIATE

```python
df[num].corr()
```

- Row ID: Tidak memiliki korelasi kuat dengan variabel lain, menunjukkan bahwa ini hanya atribut identitas.
- Postal Code: Tidak memiliki korelasi signifikan dengan variabel lain.
- Sales: Memiliki korelasi sedang dengan Profit (0.479) yang menunjukkan bahwa penjualan yang tinggi berhubungan terhadap keuntungan lebih besar.
- Quantity: Memiliki korelasi sedikit dengan Sales (0.201) menunjukkan bahwa jumlah pembelian berhubungan dengan nilai penjualan.
- Discount: Memiliki korelasi negatif sedang dengan Profit (-0.219), menunjukkan bahwa diskon yang tinggi dapat mengurangi keuntungan.
- Profit: Dipengaruhi positif oleh Sales (0.479) dan negatif oleh Discount (-0.219).

```python
# correlation heatmap
plt.figure(figsize=(8, 8))
sns.heatmap(df[num].corr(), cmap='Blues', annot=True, fmt='.2f')
```

```python
#pairplot
plt.figure(figsize=(15, 15))
sns.pairplot(df, diag_kind='kde')
```

- Sales vs. Profit: Penjualan yang tinggi cenderung menghasilkan keuntungan yang besar (Berkorelasi positif)
- Quantity vs. Sales: Peningkatan kuantitas sedikit memengaruhi nilai penjualan (Berkorelasi positif).
- Discount vs. Profit: Diskon tinggi seringkali terkait dengan keuntungan yang lebih rendah (Berkorelasi negatif).
